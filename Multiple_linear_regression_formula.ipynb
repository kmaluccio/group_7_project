```python
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
WHO_data_load = "WHO_AirQuality_Database_2018.csv"
cancer_data_to_load = "500_cancer.csv"
asthma_data_to_load = "500_asthma.csv"
diabetes_data_to_load = "500_diabetes.csv"

# Read the City and Ride Data
air_data_df = pd.read_csv(WHO_data_load)
cancer_data_df = pd.read_csv(cancer_data_to_load)
ashtma_data_df = pd.read_csv(asthma_data_to_load)
diabetes_data_df = pd.read_csv(diabetes_data_to_load)
```

```python
air_data_df
```

```python
# filter for USA data only
air_data_df = air_data_df.loc[air_data_df["country"] == "United States of America"]

# change city to CityName
air_data_df = air_data_df.rename(columns={"city":"CityName"})

air_data_df.head()
```

```python


air_cities = air_data_df["CityName"].unique()

air_cities = np.array(air_cities).tolist()

```

```python
diabetes_cities = diabetes_data_df["CityName"].unique()

diabetes_cities = np.array(diabetes_cities).tolist()

diabetes_cities
```

```python
new_cities = []

# create a for loop for each city list that checks if the city name matches
for city in air_cities:
    for i in diabetes_cities:
        if city == i:
            new_cities.append(city)


print(new_cities)
```

```python

clean_air_df = air_data_df.loc[air_data_df['CityName'].isin(new_cities)]

clean_air_df.head()
```

```python
# clean diabetes 
clean_diabetes_df = diabetes_data_df.loc[diabetes_data_df['CityName'].isin(new_cities)]

clean_diabetes_df.head()
```

```python
# Merge the datasets together

clean_air_dia_df = pd.merge(clean_air_df, clean_diabetes_df, how="left", on=["CityName","CityName"])

clean_air_dia_df.head()
```

```python
clean_air_dia_df.shape
```

```python
Clean_df = clean_air_dia_df[["Data_Value","pm10", "pm25"]].copy()
```

```python
Clean_df = Clean_df.dropna()
```

```python
# export the csv for Air_Diabetes

clean_air_dia_df.to_csv('Air_v_Diabetes_Database.csv')
```

```python
Clean_df.to_csv('Clean_df.csv')
```

```python
# Separate the Features (X) from the Target (y)
y = Clean_df["Data_Value"]
X = Clean_df[["pm10", "pm25"]]
```

```python
y.dtype
```

```python
y.isna().sum()
```

```python
#Split our data into training and testing
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    random_state=1, 
                                                    stratify=y)
X_train.shape
```

```python
#Create a Logistic Regression Model
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='lbfgs',
                                max_iter=200,
                                random_state=1)
```

```python
#Fit (train) or model using the training data

classifier.fit(X_train, y_train)
```

```python
#Make predictions
y_pred = classifier.predict(X_test)
results = pd.DataFrame({"Prediction": y_pred, "Actual": y_test}).reset_index(drop=True)
results.head(20)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred))
```
